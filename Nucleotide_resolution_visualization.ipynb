The purpose of this code is to visualize our nucleotide resolution data in the form of coverage graphs (separated or overlayed). 

```python
# import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Importing hour2 and hour6 nucleotide resolution data
hour2 = pd.read_csv('hour_2.csv')
hour6 = pd.read_csv('hour_6.csv')
```

```python
# Code to visualize hour2 coverage
fig, ax = plt.subplots(figsize= (10,6))

# Insert the range in which you would like to look over for hour 2
plt.plot(hour2['ID'].iloc[13000:14000], hour2['value'].iloc[13000:14000], color = 'blue')

ax.ticklabel_format(style = 'plain', axis='x')

plt.xlabel('Nucleotide Position')
plt.ylabel('Read Coverage (RPM)')
plt.title('Hour 2 Coverage')

plt.show()
```

```python
# Code to visualize hour 6 coverage

fig, ax = plt.subplots(figsize= (10,6))

# Insert range in which you would like to look over for hour 6
plt.plot(hour6['ID'].iloc[13000:14000], hour6['value'].iloc[13000:14000], color = 'red')

ax.ticklabel_format(style = 'plain', axis='x')

plt.xlabel('Nucleotide position')
plt.ylabel('Read Coverage (RPM)')
plt.title('Hour 6 Coverage')

plt.show()
```

```python
# Code to overlay hour 2 and hour 6 coverage data
fig,ax = plt.subplots(figsize=(8, 6))

# Insert range of interest into the two graphs
graph_h2 = plt.plot(hour2['ID'].iloc[13000:14000], hour2['value'].iloc[13000:14000], color = 'blue', alpha = 1.0)
graph_h6 = plt.plot(hour6['ID'].iloc[13000:14000], hour6['value'].iloc[13000:14000], color = 'red', alpha = 0.7)

# Add legends, titles, axes
plt.legend([graph_h2[0],graph_h6[0]], ['Hour 2 Coverage', 'Hour 6 Coverage'])
ax.ticklabel_format(style = 'plain', axis='x')

plt.xlabel('Nucleotide Position')
plt.ylabel('Read Coverage (RPM)')
plt.title('Hour 2 vs. Hour 6 Coverage')

plt.show()
```
