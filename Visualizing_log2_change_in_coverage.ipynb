This code is used to visualize the log2 change in coverage for any set of base-pairs.

```python
# import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read in log2 change in coverage data
data = pd.read_csv('Log2_change_in_coverage.csv')
```

```python
# create graph of log2 change in coverage for data
fig,ax = plt.subplots(figsize=(14, 6))

# Insert the range of interest in the line below to specify
log2_graph = plt.bar(data['ID'].iloc[13000:14000], data['log2Diff_6_minus_2'].iloc[13000:14000])
ax.ticklabel_format(style = 'plain', axis='x')

plt.xlabel('Nucleotide position')
plt.ylabel('Log base 2 Change in Coverage (Hour 6 - Hour 2)')
plt.title('Log2 Change in Coverage Between Hour 2 and Hour 6')

plt.show()
```
